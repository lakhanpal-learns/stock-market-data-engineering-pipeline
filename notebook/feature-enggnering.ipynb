# load data 

```python
import pandas as pd
import numpy as np

# Load cleaned data
data = pd.read_csv("F:\project\college project\code\data\procesed data\clean_stock_data.csv")

print(data.head())
print(data.shape)
```

# Feature enggnering

## target variable

```python
data['Target'] = data['Close'].shift(-1)
data.dropna(inplace=True)
print(data['Target'].tail() )
```

## feature1 Moving average

```python
data['MA_10'] = data['Close'].rolling(window=10).mean()
data['MA_50'] = data['Close'].rolling(window=50).mean()
```

## feaature 2 daily return

```python
data['Return'] = data['Close'].pct_change()
```

## feature 3 volatity

```python
data['Volatility'] = data['Return'].rolling(window=10).std()
```

## clean all nan value ( because of rolling function )

```python
data.dropna(inplace=True)
```
